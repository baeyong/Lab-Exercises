Top 100 Movies

```python
source = requests.get("https://www.imdb.com/list/ls055592025/").text
soup = BeautifulSoup(source, "lxml")
```

```python
section = soup.find("h3", class_="lister-item-header")
print(section.prettify())
```

```python
title = section.a.text
date = section.find("span", class_="lister-item-year text-muted unbold").text.strip("()")
title
```

```python

for element in soup.find_all("h3", class_="lister-item-header"):
    
    no = element.span.text.strip(".")
    print(no)
    
    title = element.a.text
    print(title)
    
    date = element.find("span", class_="lister-item-year text-muted unbold").text.strip("()")
    print(date)
    
 
```

```python
csv_file = open("top100movies.csv", "w", newline="", encoding="utf-8")
csv_writer = csv.writer(csv_file)
csv_writer.writerow(["No.","Title","Year of Release"])

for element in soup.find_all("h3", class_="lister-item-header"):  
 
    no = element.span.text.strip(".")
    
    title = element.a.text
    
    date = element.find("span", class_="lister-item-year text-muted unbold").text.strip("()")
    
    csv_writer.writerow([no,title,date])
    
csv_file.close()
    
```

Library of Congress Search

```python
source2 = requests.get("https://www.loc.gov/search/?q=cats&sp=1").text
soup = BeautifulSoup(source2, "lxml")
 # title, item description, webpage hyperlink

```

```python
section2 = soup.find("div", class_="item-description")
print(section2.prettify())
```

```python
title2 = section2.a.text.strip()
description = section2.find("span", class_="item-description-abstract").text.strip()
link = section2.a['href']
print(title2)
print(description)
print(link)
```

```python
csv_file = open("cats.csv", "w", newline="", encoding="utf-8")
csv_writer = csv.writer(csv_file)
csv_writer.writerow(["Title","Description","Hyperlink"])

page = 1

while page < 6:
    search = requests.get(f"https://www.loc.gov/search/?q=cats&sp={page}").text
    soup = BeautifulSoup(search, "lxml")
    
    for element in soup.find_all("div", class_="item-description"):
        title2 = element.a.text.strip()
        
        
        if element.find("span", class_="item-description-abstract"):
            description = element.find("span", class_="item-description-abstract").text.strip()
            
        
        link = element.a['href']
        
        csv_writer.writerow([title2,description,link])
    
    page += 1
    
csv_file.close()
```
