```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
PATH = Service("/Users/taeyong/Documents/School/HARP_151/Week_6_Web_Scraping/chromedriver_mac64")
driver = webdriver.Chrome(service=PATH)
```

```python
driver.get("https://www.binghamton-ny.gov/home")
```

```python
government = driver.find_element(By.XPATH, "/html/body/div[2]/div/div[2]/div[1]/div/nav/ul/li[3]/a")
departments = driver.find_element(By.XPATH, "/html/body/div[2]/div/div[2]/div[1]/div/nav/ul/li[3]/div/div/ul[1]/li/a")

actions = ActionChains(driver)
actions.move_to_element(government)
actions.click(government)
actions.click(departments)
actions.perform()
```

```python
try:
    personnel = WebDriverWait(driver,10).until(
        EC.presence_of_element_located((By.XPATH, "/html/body/div[2]/div/div[2]/div[2]/nav/ul/li/ul/li[14]/a[1]")))
    personnel.click()
    
    employment = WebDriverWait(driver,10).until(
        EC.presence_of_element_located((By.XPATH, "/html/body/div[2]/div/div[2]/div[2]/nav/ul/li/ul/li[14]/ul/li/a")))
    employment.click()
finally:
    pass
```

```python
from bs4 import BeautifulSoup
import requests
import html5lib
import csv
import pandas as pd
```

```python
source = driver.page_source
print(source)
```

```python
soup = BeautifulSoup(source, "lxml")
section = soup.find("tbody")
print(section.prettify())
```

```python
csv_file = open("bingjobs.csv", "w", newline="", encoding="utf-8")
csv_writer = csv.writer(csv_file)
csv_writer.writerow(["Job", "Type", "Deadline", "Salary"])

for element in section.find_all("tr")[1:]:
    
    job = element.find("td").text
    
    if element.find("td", attrs = {'data-th':'Type'}):
        typ = element.find("td", attrs = {'data-th':'Type'}).text

    if element.find("td", class_="mobile_hide"):
        deadline = element.find("td", class_="mobile_hide").text

    if element.find("td", attrs = {'data-th':'Salary'}):
        salary = element.find("td", attrs = {'data-th':'Salary'}).text
        
    csv_writer.writerow([job,typ,deadline,salary])

    
csv_file.close()
```
