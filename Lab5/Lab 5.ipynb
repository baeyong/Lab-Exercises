```python
try:
    main = WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.ID, "main"))
    )
    print(main.text)
    
    articles = main.find_elements(By.TAG_NAME, "article")
    
    for i in articles:
        header = i.find_element(By.CLASS_NAME, "entry-summary")
        print(header.text)
finally:
    driver.quit()
```

```python
# Lab Exercise

# use Selenium to open up the page salaries > players & scrape the players salaries. Store the data in a csv file
driver = webdriver.Chrome(service=PATH)
driver.get("https://hoopshype.com/")
link = driver.find_element(By.LINK_TEXT, 'SALARIES')
link.click()


element = WebDriverWait(driver,10).until(
    EC.presence_of_element_located((By.LINK_TEXT, 'PLAYER SALARIES'))
)
element.click()
```

```python
from bs4 import BeautifulSoup
import requests
import html5lib
import csv
import pandas as pd
```

```python
source = driver.page_source
print(source)
```

```python
soup = BeautifulSoup(source, "lxml")
section = soup.find("tr")
print(section.prettify())
```

```python
rank = section.find("td", class_="rank").text.strip("\t\n")
rank
```

```python
name = section.find("td", class_="name").text.strip("\t\n")
name
```

```python
salary = section.find("td", class_="hh-salaries-sorted").text.strip("\n\t")
salary
```

```python
csv_file = open("nbasalaries.csv", "w", newline="", encoding="utf-8")
csv_writer = csv.writer(csv_file)
csv_writer.writerow(["Rank", "Name", "Current Salary"])

for element in soup.find_all("tr", class_=""):
    
    rank = element.find("td", class_="rank").text.strip("\t\n")
    
    name = element.find("td", class_="name").text.strip("\t\n")
    
    salary = element.find("td", class_="hh-salaries-sorted").text.strip("\t\n")
    
    csv_writer.writerow([rank,name,salary])
    
csv_file.close()
    
```
